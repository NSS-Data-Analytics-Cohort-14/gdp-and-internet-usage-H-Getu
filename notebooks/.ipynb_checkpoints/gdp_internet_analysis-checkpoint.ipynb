13. Which country had the lowest GDP per capita in 2020? Create a plot showing how this country's GDP per capita has changed over the timespan of the dataset. **Bonus question:** Is it true in general that coutries had a higher GDP per capita in 2020 than in 1990? Which countries had lower GDP per capita in 2020 than in 1990?

```python
gdp_2020_min=gdp_2020[gdp_2020.GDP_Per_Capita==gdp_2020.GDP_Per_Capita.min()]

country_data_l=gdp_df[gdp_df.Country.isin(gdp_2020_min.Country)]
sns.lineplot(x='Year',y='GDP_Per_Capita',data=country_data_l)
plt.title(set(country_data_l.Country))
```

```python
all_countries=dict(gdp_df.groupby('Year')['GDP_Per_Capita'].mean())

x=list(all_countries.keys())
y=list(all_countries.values())

plt.plot(x,
         y,
         label="all_countries_mean",
        color='orange')
plt.plot(country_data_l.Year, 
         country_data_l.GDP_Per_Capita,
         label="least_gdp",
         color="blue")
plt.legend()
plt.xlabel("Year")
plt.ylabel("Gdp_per_capita")
plt.title(f'world mean gdp vs {set(country_data_l.Country)} gdp trend')
plt.show()
```

The general trend for all countries is that the mean gdp_per_Capita increases over time; whereas Burundi's gdp_per_capita decreases over time

```python
gdp_1990 = gdp_df[gdp_df['Year'] == 1990].set_index('Country')['GDP_Per_Capita']
gdp_2020 = gdp_df[gdp_df['Year'] == 2020].set_index('Country')['GDP_Per_Capita']
merged_gdp = pd.concat([gdp_1990, gdp_2020], axis=1, keys=['1990', '2020'])
countries_decline = merged_gdp[merged_gdp['1990'] > merged_gdp['2020']]
print(countries_decline)
```

17. What is the first year to have a non-zero internet users percentage value?

18. How does the distribution of internet users percent differ for 2000 and 2014?

```python
internet_df_2000 = internet_df[internet_df['Year'] == 2000].set_index('Country')['Internet_Users_Pct']
internet_df_2014 = internet_df[internet_df['Year'] == 2014].set_index('Country')['Internet_Users_Pct']
internet_combined = pd.concat([internet_df_2000,internet_df_2014], axis = 1, keys =['2000','2014'])
Avg_difference = (internet_combined['2014']-internet_combined['2000']).mean()
print(f'The Internet_Users_Pct for all countries from 2000 to 2014 increased by an avg of {Avg_difference} percent.')
```

19. For how many countries was the percentage of internet users below 5% in 2014?

```python
internet_user_2014 = internet_df[np.logical_and(internet_df['Year'] == 2014,internet_df['Internet_Users_Pct'] < 5)]
print(f'The number of countries with internet users less than 5% in 2014 is {internet_user_2014['Country'].count()}.')
```

20. Merge the two DataFrames to one. Do this in a way that keeps **all rows** from each of the two DataFrames. Call the new DataFrame `gdp_and_internet_use`. Look at the first and last few rows to confirm that it merged correctly.

```python
gdp_and_internet_use = pd.merge(gdp_df,internet_df, on =('Country', 'Year'), how = 'inner')
gdp_and_internet_use.head()
gdp_and_internet_use.tail()
```

21. Find the three countries with the highest internet users percentage in 2014. Use a seaborn FacetGrid (https://seaborn.pydata.org/generated/seaborn.FacetGrid.html) to compare how the GDP per capita has changed over time for these three countries. What do you notice?

```python
gdp_and_internet_2014 = gdp_and_internet_use[gdp_and_internet_use['Year'] == 2014]
gdp_and_internet_sorted = gdp_and_internet_2014.sort_values(by = 'Internet_Users_Pct', ascending = False)
gdp_and_internet_top3 = gdp_and_internet_sorted.head(3)
```

```python
gdp_top3 = gdp_and_internet_use[gdp_and_internet_use['Country'].isin(gdp_and_internet_top3['Country'])]
sns.lineplot(x = 'Year', y = 'GDP_Per_Capita', data = gdp_top3, hue = 'Country' )
```

24. **Stretch Question:** Use the `qcut` function from pandas (https://pandas.pydata.org/docs/reference/api/pandas.qcut.html) to divide countries in `gdp_per_capita_2014` into three groups based on their GDP per capita values. Label these groups as "Low", "Medium", and "High". Put these labels in a new column, named "GDP_group".

```python
gdp_per_capita_2014 = gdp_and_internet_use_2014.copy()
gdp_per_capita_2014["GDP_group"] = pd.qcut(gdp_per_capita_2014.GDP_Per_Capita,3,["Low", "Medium", "High"]) 
gdp_per_capita_2014.head()
```

25. **Stretch Question:** How does the median internet users percentage compare for the three gdp groups?

```python
internet_use_per_gdp = gdp_per_capita_2014.groupby("GDP_group", observed = True)["Internet_Users_Pct"].median()
print(f'The median internet users percentage for the three gdp groups is \n {internet_use_per_gdp}')
```
